### A Tale of 3 ETL's: Batch, Streaming and Job Recovery

Robust ETL solutions often entail three versions of your ETL logic:<br><br>

* The **batch** logic processes data at rest applying to a period of time or all historical data
* The **streaming** logic processes data in real time coming from data sources like Kafka 
* The **incremental** logic can be used in the case of job failure, applying to only missed data

In practice, streaming jobs that are triggered once and then halted can also solve the batch requirements of ETL jobs by maintaining metadata about which data has been processed.

Job recovery entails a number of strategies including exstensible code, monitoring, and the recovery strategy itself.

These topics will be explored in greater detail in the rest of this course.
